### Example code using the polyfit and kfold functions

**Note: This section is not part of the homework problem, but provides some potentially-helpful example code regarding the usage of numpy.polyfit, numpy.polyval, and sklearn.model_selection.KFold.**

First, let's generate some synthetic data: a quadratic function plus some Gaussian noise.

```python
# Coefficients of the quadratic function, y(x) = ax^2 + bx + c:
a = 2
b = 5
c = 7

N = 200       # Number of data points
x = np.linspace(-10, 10, num = N)                # x ranges from -10 to 10
# y is the quadratic function of x specified by a, b, and c, plus noise
y = a*x**2 + b*x + c + 15* np.random.randn(N)

# Plot the data:
plt.figure()
plt.plot(x, y, marker = 'o', linewidth = 0)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic data')

plt.show()
```

Next, we'll use the numpy.polyfit function to fit a quadratic polynomial to this data. We can evaluate the resulting polynomial at arbitrary points.



```python
# Fit a degree-2 polynomial to the data:
degree = 2
coefficients = np.polyfit(x, y, degree)

# Print out the resulting quadratic function:
print('We fit the following quadratic function: f(x) = %f*x^2 + %f*x + %f' %  \
      (coefficients[0], coefficients[1], coefficients[2]))

# Evaluate the fitted polynomial at x = 4:
x_test = 4
f_eval = np.polyval(coefficients, x_test)
print('\nf(%i) = %f' % (x_test, f_eval))

# Let's visualize our fitted quadratic:
plt.figure()

plt.plot(x, y, marker = 'o', linewidth = 0)
plt.plot(x, np.polyval(coefficients, x), color = 'red', linewidth = 3)

plt.legend(['Data', 'Fitted quadratic'], loc = 'best')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Synthetic data with Fitted Quadratic')

plt.show()
```

# Loading the Data for Problem 2

This code loads the data from bv_data.csv using the load_data helper function. Note that data[:, 0] is an array of all the x values in the data and data[:, 1] is an array of the corresponding y values.


```python
def load_data(filename):
    """
    Function loads data stored in the file filename and returns it as a numpy ndarray.
    Input:
        filename: given as a string.
    Output:
        Data contained in the file, returned as a numpy ndarray
    """
    return np.loadtxt(filename, skiprows=1, delimiter=',')
```

```python
data = load_data('data/bv_data.csv')
x = data[:, 0]
y = data[:, 1]
```

Write your code below for solving problem 2 part B:

```python
# degrees of polynomial
d = np.array([1, 2, 6, 12])

# Number of samples used for  
N = np.arange(20, 101, 5)

# k-fold 
num_folds = 5
kf = KFold(n_splits=num_folds)

DEBUG = False
PLOT = True
```

```python

for degree in d:
    if DEBUG: print('degree: ', degree)
    
    # Containers for plotting
    TEST_MSE_DEGREE = []
    TRAIN_MSE_DEGREE = []
    for data_size in N:
        if DEBUG: print('data_size: ', data_size)
    
        i = 1
        # Iterate through cross-validation folds:
        for train_index, test_index in kf.split(x[:data_size]):
                        
            # Training and testing data points for this fold:
            x_train, x_test = x[train_index], x[test_index]
            y_train, y_test = y[train_index], y[test_index]
            
            # Print out test indices:
            if DEBUG: print('Fold ', i, ' of ', num_folds, ' test indices:', test_index)
            i += 1

            # Fit a degree polynomial to the data:
            coefficients = np.polyfit(x_train, y_train, degree)

            # Evaluate the fitted polynomial
            f_eval_test = np.polyval(coefficients, x_test)
            test_mse = np.mean((f_eval_test-y_test)**2)
            if DEBUG: print("Mean validation error: ", test_mse)

            f_eval_train = np.polyval(coefficients, x_train)
            train_mse = np.mean((f_eval_train-y_train)**2)
            if DEBUG: print("Mean validation error: ", train_mse)

        TEST_MSE_DEGREE.append(test_mse)
        TRAIN_MSE_DEGREE.append(train_mse)

    # visualize test mse for polyfit degree
    if PLOT:
        plt.figure()
        plt.plot(N, TEST_MSE_DEGREE)
        plt.plot(N, TRAIN_MSE_DEGREE)
        plt.xlabel('Data Size')
        plt.ylabel('MSE')
        plt.legend(['Validation Error', 'Training Error'], loc = 'best')
        title = 'Polyfit for degree ' + str(degree)
        plt.title(title)
        plt.axis([20, 100, 0, 6])
        plt.grid(True)
        plt.show()

        plt.figure()
        # plt.subplot(1, 2, 2)
        plt.plot(x, y, marker = 'o', linewidth = 0)
        x_sorted = np.linspace(min(x), max(x), 100)
        plt.plot(x_sorted, np.polyval(coefficients, x_sorted), color = 'red', linewidth = 3)

        plt.legend(['Data', 'Fitted quadratic'], loc = 'best')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Synthetic data with Fitted Quadratic')

        plt.show()
    
```

### Observations
1. Model error decreases as we get more data
2. As the model complexity increases, training error decreases
3. For linear model, both Training Error & Validation Error are highest => Model is Underfitting
4. For too complex model (degree 12), Training Error >>> Validation Error => Model is Overfitting
